```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
path_to_ds = "D:\\SomeMachineLearningAlgorithm\\SomeMachineLearningAlgorithm\\avocadoPrice\\avocado.csv"
datas = pd.read_csv(path_to_ds)
datas.head()
```

```python
datas = datas.drop(["Unnamed: 0", "Date", "XLarge Bags", "type", "year", "region"], axis=1)
datas.head()
```

```python
datas_features = datas.copy()
datas_features.head()
```

```python
datas_label = datas_features.pop('AveragePrice')
```

```python
datas_features = np.array(datas_features)
datas_features
```

```python
normalize = layers.Normalization()
```

```python
normalize.adapt(datas_features)
```

```python
model = tf.keras.Sequential([
    layers.Dense(16, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])
```

```python
model.compile(
    optimizer=tf.optimizers.Adam(),
    metrics=['accuracy'],
    loss=tf.losses.MeanSquaredError()
)
```

```python
model.fit(datas_features, datas_label, epochs=100)
```

```python
X_new = np.array([[64236.62, 850.74, 54454.85, 48.16, 8696.87, 8103.62, 93.25]])
```

```python
prediction = model.predict(X_new)
print(f"Prediction: {prediction}")
```
